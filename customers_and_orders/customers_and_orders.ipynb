**Question 2:** In how many different states do the customers live in?

```python
# open file and init vars
# evaluate the contents of the file this follows the same exact strategy as in
# counting the customers, since we need to count each state just once it suits
# to use a set again
states = set()
with open(datafile, 'r') as fl:
  csv_reader = csv.DictReader(fl)
  for row in csv_reader:
    states.add(row['State'])
#for states list visual, let us print the set

'''print(sorted(states))'''

# despite uniqueness of set elements, the format in which California state is
# being stored causes problems, let's normalize this

states_normalized_set = {state.strip().upper() for state in states}

# this solution should be used at the moment of populating data to the set, but
# for the sake of the challenge's evaluation I have kept the whole train of
# thought I followed

'''print(sorted(states_normalized_set))'''

# there we go!, now that we know our states sets is ok, we can answer the
# question
states_number = len(states_normalized_set)
print('There are {} states in which customers live.'.format(states_number))
```

**Question 3** What is the state with most customers?

```python
# here we will use a dictionary, in which the keys will be the normalized-format
# states the content of each key, the number of customer from said state, pretty
# straight forward
customers_per_state = {}
with open(datafile,'r') as fl:
  csv_reader = csv.DictReader(fl)

  for row in csv_reader:
    state = row['State'].strip().upper()

    if state in customers_per_state:
      customers_per_state[state] += 1
    else:
      customers_per_state[state] = 1

state_max_cx = max(customers_per_state, key = customers_per_state.get)
print('The state with the most customers is {} with {} customers.'.format(state_max_cx, customers_per_state[state_max_cx]))
```

**Question 4** What is the state with the least customers?

```python
# same solution as before, but we will use min insteand of max and we will use
# the same dictionary from Q3
state_min_cx = min(customers_per_state, key = customers_per_state.get)
print('The state with the least customers is {} with {} customers.'.format(state_min_cx, customers_per_state[state_min_cx]))
```

**Question 5:** What is the most common last name?

```python
#this will combine the logic from all previous questions

#first a dictionary, keys will be the last nanes and content will be the last
#names occurences

last_names = {}

with open(datafile, 'r') as fl:
  csv_reader = csv.DictReader(fl)

  #same logic for data population

  for row in csv_reader:
    last_name = row['LastName'].strip().upper()

    if last_name in last_names:
      last_names[last_name] += 1
    else:
      last_names[last_name] = 1

most_common_last_name = max(last_names, key = last_names.get)
print('The most common last name is {} with {} customers having it as their last name.'.format(most_common_last_name, last_names[most_common_last_name]))
```
